```python
library = []

def add_book():
    title = input("Enter the title of the book: ")
    author = input("Enter the Author of the book: ")
    isbn = input("Enter the ISBN of the book: ")
    book = {
        "title": title,
        "author": author,
        "isbn": isbn
    }
    library.append(book)
    print("Book added successfully")

def remove_book():
    isbn = input("Enter the ISBN of the book to remove: ")
    for book in library:
        if book['isbn'] == isbn:
            library.remove(book)
            print("Book removed successfully")
            return
    print("Book not found")

def view_books():
    if not library:
        print("No books available in the system")
    else:
        for book in library:
            print(f"Title: {book['title']}, Author: {book['author']}, ISBN: {book['isbn']}")

def search_book():
    keyword = input("Enter title or author to search: ").lower()
    found_books = []
    for book in library:
        if keyword in book['title'].lower() or keyword in book['author'].lower():
            found_books.append(book)
    if found_books:
        print("Books found:")
        for book in found_books:
            print(f"Title: {book['title']}, Author: {book['author']}, ISBN: {book['isbn']}")
    else:
        print("No books found")

def menu():
    while True:
        print("\nWelcome to book management system")
        print("1. Add a book")
        print("2. Remove a book")
        print("3. View all books")
        print("4. Search for a book")
        print("5. Exit")
        
        choice = input("Enter your choice (1-5): ")
        if choice == '1':
            add_book()
        elif choice == '2':
            remove_book()
        elif choice == '3':
            view_books()
        elif choice == '4':
            search_book()
        elif choice == '5':
            print("Thank you for using book management system. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

menu()
    
```

```python
import pandas as pd

library = []
df = pd.DataFrame(library)

def add_book():
    title = input("Enter the title of the book: ")
    author = input("Enter the Author of the book: ")
    isbn = input("Enter the ISBN of the book: ")
    book = {
        "title": title,
        "author": author,
        "isbn": isbn
    }
    library.append(book)
    global df
    df = pd.DataFrame(library)
    print("Book added successfully")

def remove_book():
    isbn = input("Enter the ISBN of the book to remove: ")
    global df
    for book in library:
        if book['isbn'] == isbn:
            library.remove(book)
            df = pd.DataFrame(library)
            print("Book removed successfully")
            return
    print("Book not found")

def view_books():
    if df.empty:
        print("No books available in the system")
    else:
        print(df.to_string(index=False))

def search_book():
    isbn = input("Enter the ISBN of the book to search: ")
    result = df[df['isbn'] == isbn]
    if result.empty:
        print("No such book found!")
    else:
        print(result.to_string(index=False))

def menu():
    while True:
        print("\nWelcome to book management system")
        print("1. Add a book")
        print("2. Remove a book")
        print("3. View all books")
        print("4. Search for a book")
        print("5. Exit")
        
        choice = input("Enter your choice (1-5): ")
        if choice == '1':
            add_book()
        elif choice == '2':
            remove_book()
        elif choice == '3':
            view_books()
        elif choice == '4':
            search_book()
        elif choice == '5':
            print("Thank you for using book management system. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

menu()
```
