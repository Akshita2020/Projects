### Filling Missing Values in Postal Code by City, State, and Country Columns Using Group Mode

```python
# The code groups the data by city,state,country and for each city,state and country it fills in missing postal codes.
# It uses the most common postal code (if available) to replace any missing values.
# If there is no common postal code (i.e., all values are missing or different), it marks those missing values with pd.NA.
```

```python
superstore['Postal Code'] = superstore.groupby('City')['Postal Code'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else pd.NA)
)
```

```python
superstore['Postal Code'] = superstore.groupby('State')['Postal Code'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else pd.NA)
)
```

```python
superstore['Postal Code'] = superstore.groupby('Country')['Postal Code'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else pd.NA)
)
```

```python
superstore.isnull().sum()
```

### Conclusion:
The missing values in the Postal Code column have been addressed by filling 17,635 of the 41,296 missing entries. The remaining 17,661 entries could not be filled because:

The code groups the data by City, State, and Country and fills in missing Postal Codes for each group.
It uses the most common Postal Code (if available) to replace any missing values.
If there is no common Postal Code (i.e., all values are missing or different), those missing values are marked as pd.NA.

# Multivariate Analysis 

## What are the average sales and profit across categories and segments after removing outliers?

```python

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]


filtered_superstore = remove_outliers(superstore, 'Sales')
filtered_superstore = remove_outliers(filtered_superstore, 'Profit')

plt.figure(figsize=(14, 6))
sns.boxplot(data=filtered_superstore, x='Category', y='Sales', hue='Segment', dodge=True)
plt.title('Sales Distribution by Category and Segment (Outliers Removed)', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.grid(True)
plt.legend(title='Segment')
plt.savefig("Sales Distribution by Category and Segment (Outliers Removed)")
plt.show()

plt.figure(figsize=(14, 6))
sns.boxplot(data=filtered_superstore, x='Category', y='Profit', hue='Segment', dodge=True)
plt.title('Profit Distribution by Category and Segment (Outliers Removed)', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Profit', fontsize=14)
plt.grid(True)
plt.legend(title='Segment')
plt.savefig("Profit Distribution by Category and Segment (Outliers Removed)")
plt.show()
```

### Conclusion - 
Technology category leads: It has the highest median sales and profits across all segments.
    
Furniture underperforms: It shows lower sales and profits compared to other categories.
    
Segment similarities: Within each category, the distributions across segments (Corporate, Consumer, Home Office) are relatively similar.

Profit variability: Office Supplies and Furniture show more variability in profits, including some negative values.
    
Sales consistency: Sales distributions are more consistent across categories than profit distributions.
    
Outliers impact: Removing outliers provides a clearer view of typical performance, but extreme values still exist, especially in sales.

Category hierarchy: Overall performance ranks as Technology > Office Supplies > Furniture in both sales and profits.

 ## What are the average sales for each of the top 5 postal codes, broken down by customer segment?

```python
# Filter the data for the top 5 postal codes to avoid clutter
top_postal_codes = superstore['Postal Code'].value_counts().head(5).index
filtered_data = superstore[superstore['Postal Code'].isin(top_postal_codes)]

# Calculate average sales by Postal Code and Segment
avg_sales_by_segment = filtered_data.groupby(['Postal Code', 'Segment'])['Sales'].mean().reset_index()

plt.figure(figsize=(14, 6))
sns.barplot(data=avg_sales_by_segment, x='Postal Code', y='Sales', hue='Segment', estimator=np.mean)
plt.title('Average Sales by Postal Code and Segment (Top 5)', fontsize=16)
plt.xticks(rotation=45)
plt.xlabel('Postal Code', fontsize=14)
plt.ylabel('Average Sales', fontsize=14)
plt.grid(True)
plt.legend(title='Segment')
plt.savefig("Average Sales by Postal Code and Segment (Top 5)")
plt.show()
```

### Conclusion -
Sales vary significantly across postal codes and segments

Home office segment dominates in postal code 10024.0

No segment consistently leads across all postal codes

Corporate segment generally underperforms

Postal codes 10024.0 and 10009.0 show strongest overall sales

Postal code 94122.0 has a unique pattern with low home office sales
